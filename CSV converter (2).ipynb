```python
import csv
import random
def csvFormat1(sourcedoc):
    with open (sourcedoc, "r", encoding="utf-8") as f:
        reader=csv.reader(f)
        next(reader) # Skip the first row (headings)
        last = "blank" #Previous row
        count = 0 #Number of questions for current film clip
        count2 = 0 #Number of film clips
        data = [] #Film clips array
        data2 = [] #Quiz question info array
        counts = [] #Stores number of questions for each film clip
        choices = [] #Answer choices array
        correctAnswers = [] #Indices of correct answers
        for row in reader:
            if row[0]!=last: #First cell of row doesn't match with that of previous row - signals a new film clip
                data.append(row[0]) #Clip title
                data.append(row[1]) #Embed link
                count2 = count2 + 1 #Increment number of film clips
                counts.append(count) #Add number of questions to array
                count = 0 #Reset number of questions for next clip
            data2.append(row[2]) #Question text
            choices.append(row[3]) #Answer choices
            choices.append(row[4])
            choices.append(row[5])
            choices.append(row[6])
            random.shuffle(choices) #Randomize answer choices
            for i in range(4):
                data2.append(choices[i]) #Add answer choices to question info array
                if choices[i] == row[3]: #If choice matches correct answer:
                    correctAnswers.append(i) #Save index in array
            choices = [] #Empty choices array for next question
            last=row[0] #Update previous row
            count = count + 1 #Increment number of questions
        counts.remove(0) #Remove first element from array
        counts.append(count) #Add number of questions for last clip to array
        with open ("OutputCSV.csv", "w", encoding="utf-8") as g:
            writer=csv.writer(g)
            place = 0 #Current index of film clips array
            place2 = 0 #Current index of question info array
            place3 = 0 #Index of current question (in correct answers array)
            for x in range (0,count2): #For the number of clips:
                g.write("\n")
                g.write("clip-title,")
                if "," in data[place]: #Check for commas 
                    g.write('"' + data[place] + '"') #Surround with quotes
                else:
                    g.write(data[place])
                    """
                    If a clip title, question text, or answer choice contains a comma,
                    the code will surround it with quotation marks so that
                    the commas do not separate the string into separate cells.
                    """
                g.write("\ndescription")
                g.write("\nthumbnail")
                g.write("\nimg")
                g.write("\nsrc,")
                g.write(data[place+1])
                g.write("\nclip-tags")
                g.write("\n")
                place=place+2 #Update current index of clips array
                for y in range (0,counts[x]): #For the number of questions for the current clip:
                    g.write("\n")
                    g.write("question-text,")
                    if "," in data2[place2]: #Check for commas
                        g.write('"' + data2[place2] + '"\n') #Surround with quotes
                    else:
                        g.write(data2[place2]+"\n")
                    for z in range(4): #The next 4 elements are the (randomized) answer choices.
                        if z == correctAnswers[place3]: #If z matches this question's correct answer index:
                            g.write("correct,")
                        else:
                            g.write("incorrect,")
                        if "," in data2[place2+z+1]: #Check for commas
                            g.write('"' + data2[place2+z+1] + '"\n') #Surround with quotes
                        else:
                            g.write(data2[place2+z+1]+"\n")
                    g.write("question-tags")
                    g.write("\n")
                    place2=place2+5 #Update current index of question info array
                    place3=place3+1 #Update index of current question (correct answers array)
                g.write("\n")
                g.write("end of quiz")
                g.write("\n")
```

```python
#All comments from the previous code apply to this code as well.
import csv
import random
def csvFormat2(sourcedoc):
    with open (sourcedoc, "r", encoding="utf-8") as f:
        reader=csv.reader(f)
        next(reader)
        last = "blank"
        count = 0
        count2 = 0
        data = []
        data2 = []
        counts = []
        choices = []
        correctAnswers = []
        for row in reader:
            if row[0]!=last:
                data.append(row[1])
                data.append(row[11])
                count2 = count2 + 1
                counts.append(count)
                count = 0
            data2.append(row[6])
            choices.append(row[7])
            choices.append(row[8])
            choices.append(row[9])
            choices.append(row[10])
            random.shuffle(choices)
            for i in range(4):
                data2.append(choices[i])
                if choices[i] == row[7]:
                    correctAnswers.append(i)
            choices = []
            last=row[0]
            count = count + 1
        counts.remove(0)
        counts.append(count)
        with open ("OutputCSV2.csv", "w", encoding="utf-8") as g:
            writer=csv.writer(g)
            place = 0
            place2 = 0
            place3 = 0
            for x in range (0,count2):
                g.write("\n")
                g.write("clip-title,")
                if "," in data[place]:    
                    g.write('"' + data[place] + '"')
                else:
                    g.write(data[place])
                g.write("\ndescription")
                g.write("\nthumbnail")
                g.write("\nimg")
                g.write("\nsrc,")
                g.write(data[place+1])
                g.write("\nclip-tags")
                g.write("\n")
                place=place+2
                for y in range (0,counts[x]):
                    g.write("\n")
                    g.write("question-text,")
                    if "," in data2[place2]:
                        g.write('"' + data2[place2] + '"\n')
                    else:
                        g.write(data2[place2]+"\n")
                    for z in range(4):
                        if z == correctAnswers[place3]:
                            g.write("correct,")
                        else:
                            g.write("incorrect,")
                        if "," in data2[place2+z+1]:
                            g.write('"' + data2[place2+z+1] + '"\n')
                        else:
                            g.write(data2[place2+z+1]+"\n")
                    g.write("question-tags")
                    g.write("\n")
                    place2=place2+5
                    place3=place3+1
                g.write("\n")
                g.write("end of quiz")
                g.write("\n")
```
